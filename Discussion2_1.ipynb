```python
import os

file_path1 = '/content/drive/MyDrive/CPSC_483/GDP_Example/Happiness.csv'
file_path2 = '/content/drive/MyDrive/CPSC_483/GDP_Example/WEOOct2023all.xls'

if os.path.exists(file_path1) and os.path.exists(file_path2):
  print("Files exist.")
else:
  print("One or both files do not exist.")
```

```python
import numpy as np
import pandas as pd
import ipywidgets as widgets

hls_all_raw = pd.read_csv(file_path1)
print(hls_all_raw)
```

```python
print(hls_all_raw["Indicator"])
print("\n===========================================================\n")
print(hls_all_raw["Indicator"].unique())
print("\n===========================================================\n")
print(hls_all_raw["Type of indicator"].unique())
print("\n===========================================================\n")
hls_slice = pd.DataFrame(hls_all_raw, columns =["Country","Indicator","Type of indicator","Time","Value"])
print(hls_slice)
```

```python
hls_ls = hls_slice.loc[hls_all_raw["Indicator"] == "Long-term unemployment rate"]
print(hls_ls)
print("\n===========================================================\n")
print("Total records:")
print(len(hls_ls))

print("\n===========================================================\n")
print("Total Unique Countries:")
print(len(hls_ls["Country"].unique()))

print("\n===========================================================\n")
print("Country List")
print(hls_ls["Country"].unique())
```

```python
hls_train = hls_ls.loc[hls_ls["Time"] == 2018]
hls_train = hls_train.loc[hls_ls["Type of indicator"] == "Deprivation"]
print("\n===========================================================\n")
print("Total records:")
print(len(hls_train))

print("\n===========================================================\n")
print("Total Unique Countries:")
print(len(hls_train["Country"].unique()))

print("\n===========================================================\n")
print("Record:")
print(hls_train)
```

```python
weo_raw = pd.read_csv(file_path2, sep='\t', encoding='latin1')
print(weo_raw)
```

```python
unique_values = weo_raw["WEO Subject Code"].unique()
print(unique_values)
```

```python
weo_selected_measurement = weo_raw.loc[weo_raw['WEO Subject Code'].str.contains("NGDP_RPCH", na=False)]
weo_selected_measurement_2018 = pd.DataFrame(weo_selected_measurement, columns=['Country', '2018'])

print(weo_selected_measurement_2018)
```

```python
merged_train_data = pd.merge(hls_train, weo_selected_measurement_2018, on="Country")
merged_train_data = merged_train_data.rename(columns={"Value": "Long-term unemployment rate", "2018": "Gross domestic product"})
merged_train_data = pd.DataFrame(merged_train_data, columns=['Country','Long-term unemployment rate', 'Gross domestic product'])

print(merged_train_data)
```

```python
import matplotlib.pyplot as plt
import sklearn.linear_model
from sklearn.metrics import mean_absolute_error



X = np.c_[merged_train_data["Gross domestic product"]]
Y = np.c_[merged_train_data["Long-term unemployment rate"]]
x = X.tolist()
y = Y.tolist()

# plot data
out1 = widgets.Output()
with out1:
  plt.scatter(x, y)
  plt.xlabel('Gross domestic product')
  plt.ylabel('Long-term unemployment rate')
  plt.title("Data Plot")
  plt.show()

# fit linear model
model = sklearn.linear_model.LinearRegression()
model.fit(X, Y)

# plot predictions
predict_x = [x for x in range(901)]
predict_x = [[x/100] for x in predict_x]
predict_y = model.predict(predict_x)

out2 = widgets.Output()
with out2:
  plt.scatter(predict_x, predict_y)
  plt.scatter(x, y)
  plt.xlabel('Gross domestic product')
  plt.ylabel('Long-term unemployment rate')
  plt.title("Prediction Line")
  plt.show()

display(widgets.HBox([out1,out2]))

predictions = model.predict(X)

mae = mean_absolute_error(Y, predictions)
print("Mean Absolute Error: ", mae)
```
